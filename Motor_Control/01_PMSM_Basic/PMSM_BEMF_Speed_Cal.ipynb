## From PMSM 3 phase voltage (sin PWM modulation)

```python
#Verfiy Infineon TLE9879 BEMF phase voltage compare
#BEMF is sin/cos
import numpy as np
import matplotlib.pyplot as plt
 
freq=0.5
omeg=2*np.pi*freq
amplitude=1
step=0.001
t=np.arange(0,np.pi,step)
t_length = len(t)
compare_result=np.arange(0,np.pi,step)#
#BEMF is sin/cos
U=amplitude*np.cos(2*np.pi*freq*t)+amplitude
V=amplitude*np.cos(2*np.pi*freq*t-2*np.pi/3)+amplitude
W=amplitude*np.cos(2*np.pi*freq*t+2*np.pi/3)+amplitude
#BEMF is SVPWM
# U=amplitude*np.sin(2*np.pi*freq*t)+amplitude+ amplitude*np.sin(2*np.pi*freq*t/3)
# V=amplitude*np.sin(2*np.pi*freq*t+2*np.pi/3)+amplitude+amplitude*np.sin(2*np.pi*freq*t/3+2*np.pi/3)
# W=amplitude*np.sin(2*np.pi*freq*t-2*np.pi/3)+amplitude+amplitude*np.sin(2*np.pi*freq*t/3-2*np.pi/3)

compare=(V+W)/2


for time in range(0, t_length):
    if U[time]<compare[time]:
        compare_result[time] =0
    else:
        compare_result[time] =1


fig = plt.figure(figsize=(20, 10), dpi=80)

plt.plot(t,U,label='u')
plt.plot(t,V,linestyle='--',label='V')
plt.plot(t,W,linestyle='--',label='W')
plt.xlabel('time')
plt.ylabel('Y')
plt.title('U V W phase voltage reverse clockwise spin')
plt.legend()
plt.grid(axis='x')
plt.show()

fig = plt.figure(figsize=(20, 10), dpi=80)
plt.plot(t,U,label='u')
plt.plot(t,compare,linestyle='--',label='com (V+W)/2')
plt.plot(t,compare_result,linestyle='--',label='cmp_res')
plt.xlabel('time')
plt.ylabel('Y')
plt.title('U (V+W)/2 and compare result')
plt.legend()
plt.grid(axis='x')
plt.show()

```

## From PMSM 3 phase voltage (SVPWM modulation)

```python
#Another way to calculate motor speed base on ADC sense the BEMF voltage
#Compare U+V and U+W, we can calculate the periode of U
import numpy as np
import matplotlib.pyplot as plt
 
freq=0.5
omeg=2*np.pi*freq
amplitude=1
step=0.001
t=np.arange(0,np.pi,step)
t_length = len(t)
compare_result=np.arange(0,np.pi,step)
U=amplitude*np.sin(2*np.pi*freq*t)+amplitude
V=amplitude*np.sin(2*np.pi*freq*t+2*np.pi/3)+amplitude
W=amplitude*np.sin(2*np.pi*freq*t-2*np.pi/3)+amplitude
compare_1=U+V
compare_2=U+W

for time in range(0, t_length):
    if compare_1[time]<compare_2[time]:
        compare_result[time] =0
    else:
        compare_result[time] =1

fig = plt.figure(figsize=(20, 10), dpi=80)
plt.plot(t,U,label='u')
plt.plot(t,V,linestyle='--',label='V')
plt.plot(t,W,linestyle='--',label='W')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('U V W phase voltage')
plt.legend()
plt.grid(axis='x')
plt.show()
fig = plt.figure(figsize=(20, 10), dpi=80)

plt.plot(t,U,label='u')

plt.plot(t,compare_1,linestyle='--',label='com_1 U+V')
plt.plot(t,compare_2,linestyle='--',label='cmp_2 U+W')
plt.plot(t,compare_result,linestyle='--',label='compare_result')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('U (V+W)/2 and compare result')
plt.legend()
plt.grid(axis='x')
plt.show()
```
