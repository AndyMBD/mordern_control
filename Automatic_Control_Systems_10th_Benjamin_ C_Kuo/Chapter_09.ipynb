```python
import numpy as np
R=1;L=1;C=1
t=np.arange(0,30,0.02)
num1=[1]
den1=[L*C,R*C,1]
num2=[C,0]
den2=[L*C,R*C,1]
G1=ct.tf(num1,den1)
G2=ct.tf(num2,den2)
time,G1_step=ct.step_response(G1,t)
time,G2_step=ct.step_response(G2,t)
plt.plot(time,G1_step,'--',label='G1')
plt.plot(time,G2_step,'-.',label='G2')
plt.legend()
```
