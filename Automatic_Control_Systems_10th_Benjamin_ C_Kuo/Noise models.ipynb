# Noise models

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy
```

```python
signal = numpy.random.randn(100)
```

```python
plt.plot(signal)
```

```python
import scipy.signal
```

```python
result = []
```

```python
previous = 0
```

```python
alpha = 0.95
```

```python
for s in signal:
    news = previous*(1-alpha) + alpha*s
    previous = news
    result.append(news)
```

```python
plt.plot(result)
```
