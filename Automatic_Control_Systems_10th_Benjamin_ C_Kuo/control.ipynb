## Laplace Transfer

```python
t, s = sympy.symbols('t, s')
a = sympy.symbols('a', real=True, positive=True)
f = sympy.exp(-a*t)
f
l_f=sympy.laplace_transform(f,t,s)
l_f
```

```python
t,t0,s = sympy.symbols('t,t0, s')
a = sympy.symbols('a', real=True, positive=True)
f = sympy.exp(-a*(t-t0))
f
l_f=sympy.laplace_transform(f,t,s)
l_f
```

```python
t,t0,s = sympy.symbols('t,t0, s')
a = sympy.symbols('a', real=True, positive=True)
l_f1=1/(s+1)
l_f2=1/(s+2)
l_f3=l_f1*l_f2
l_f1
l_f2
l_f3
f1=sympy.inverse_laplace_transform(l_f1,s,t)
f2=sympy.inverse_laplace_transform(l_f2,s,t)
f3=sympy.inverse_laplace_transform(l_f3,s,t)
f1
f2
f3
```

## Inverse Laplace Transfer

```python
t,t0,s = sympy.symbols('t,t0, s')
a = sympy.symbols('a', real=True, positive=True)
l_f=1/(s+a)
f=sympy.inverse_laplace_transform(l_f,s,t)
f
```

## Control analysis transfer funciton

```python
s=ct.TransferFunction.s
tf1=1/(s+1)**2
tf1
```

### Bode Plot

```python
mag,phase,omega=ct.bode_plot(tf1)
```

```python
roots,gains=ct.root_locus(tf1,print_gain=False)
```

## Nichols Plot

```python
ct.nichols_plot(tf1, omega=None, grid=None)
```

### Zero Poles Plot

```python
ct.tfdata(tf1)
zeros,poles=ct.pzmap(tf1)
```

### Step Plot

```python
t,y=ct.step_response(tf1)
plt.plot(t, y)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Step response for G')
plt.grid()
```
