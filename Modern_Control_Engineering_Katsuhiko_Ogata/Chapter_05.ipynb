### Unit-Ramp Response of First-Order Systems.

```python
a,t, s = sympy.symbols('a,t, s')
T = sympy.symbols('T', real=True, positive=True)
t = sympy.symbols('t', real=True, positive=True)
C_S=1/(T*s+1)
R_S=1/s**2
display(C_S)
display(R_S)
display(C_S*R_S)
C_t=sympy.inverse_laplace_transform(C_S*R_S, s, t)
display(C_t)
```

```python
R_t=t
e_t=R_t-C_t
display(e_t)
```

### Unit-Impulse Response of First-Order Systems

 $C(s)=\frac{1}{Ts+1}\frac{1}{s}$

```python
a,t, s = sympy.symbols('a,t, s')
T = sympy.symbols('T', real=True, positive=True)
t = sympy.symbols('t', real=True, positive=True)
C_S=1/(T*s+1)
R_S=1
display(C_S)
display(R_S)
display(C_S*R_S)
C_t=sympy.inverse_laplace_transform(C_S*R_S, s, t)
display(C_t)
```
