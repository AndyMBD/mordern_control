# 3.1 The s-plane

## 3.1.1 The general form of all exponential and sin/cos waves

Let's review sin/cos waves  
The general equation for a sin/cos wave is two exponentials added together  
  
$$f(t)=\frac{A}{2}e^{st}+\frac{A^*}{2}e^{s^*t}$$  
where $A$ is a phasor and $s$ is the frequency of the wave 
  
When $A=-jA_m$ and $s=j\omega t$ in [radians/second], the wave is a sin wave  
When $A=A_m$ and $s=j\omega t$ in [radians/second], the wave is a cos wave  
When the phasor $A$ is a complex number of the form $A_m\angle\phi$, the wave is a general phase shifted sin/cos wave  
  
If $A$ is a real number, $A=A_m$, and $s$ is a real number, $s=\alpha$ in [Nepers/second]  
  
then our general form of the sin/cos equation reduces to a single exponential  
Because $A_m$ and $\alpha$ are real numbers $A_m^*=A_m$ and $\alpha^*=\alpha$
  
$$f(t)=\frac{A_m}{2}e^{\alpha t}+\frac{A_m}{2}e^{\alpha t}$$  
which simplifies to  
  
$$f(t)=A_me^{\alpha t}$$  
so the general form of the sin/cos equation is also the general form of a single exponential  
  
What this means, is that the expression  
  
$$f(t)=\frac{A}{2}e^{st}+\frac{A^*}{2}e^{s^*t}$$  
describes all growing or decaying exponentials and sin/cos waves! 

## 3.1.2 Review of what is a phasor and what is a phasor for real exponential functions?

Recall the complex plane

The phasor is a vector on the complex plane, and is a "snapshot" of the sin/cos wave at $t=0$  
For example, the phasor for a sin/cos wave with an amplitude of $5$ and a phase shift of $53.1^\circ$  
is $5\;\angle53.1^\circ=3+4j$
![image2!](./images_s3/image_3_1_2.png)  

The time dependent portion of the two exponential terms, $e^{j\omega t}$ and $e^{-j\omega t}$,  
causes the two phasors to spin in opposite directions.  
When these two phasors are added together they create a sin/cos function  
![image3!](./images_s3/image_3_1_3.png)  

A real exponential function of the form  

$$f(t)=\frac{A_m}{2}e^{\alpha t}+\frac{A_m}{2}e^{\alpha t}=A_me^{\alpha t}$$  
has two Phasors multipled by time dependent terms and added together, just like sin/cos waves do  
  
It's the imaginary exponential that causes the phasor to spin.  
Here are a few examples for $e^{jt}$, when $t=0, 1, 2,$ and $3$  
  
$e^{j0}=1+0j$, $e^{j1}=0.54+0.84j$, $e^{j2}=-0.42+0.91j$, and $e^{j3}=-0.99+0.14j$, 
  
However, for real exponentials the time dependent part of the exponentials is not imaginary  
Look at these examples for $e^{t}$, when $t=0, 1, 2,$ and $3$  
  
$e^{0}=1+0j$, $e^{1}=2.72+0j$, $e^{2}=7.39+0j$, and $e^{3}=20.09+0j$,  
  
The phasor doesn't spin, it just stays on the real axis  
In this example, the phasor is growing, because $\alpha=+1$

## 3.1.3 The frequency s is really a complex frequency

We are now in a position to investigate complex frequencies, $s=\alpha\pm j\omega$  
We have seen when $\omega=0$ ($s=\alpha$), the wave is either a growing or shrinking exponential, depending on the sign of $\alpha$  
On the other hand, when $\alpha=0$ and $s=\pm j\omega$, the wave is a sin/cos wave  
We have even seen that when $\alpha=0$ and $\omega=0$; ($s=0$), the wave is a constant DC signal.  
  
So the question is, what if $\alpha$ and $\omega$ are both not zero?  
  
As a reminder, it's always easier to use the exponential form for math,  
but to show what the complex frequency does, we will convert the exponential form to the trigonometric form  
  
Starting from  
$$f(t)=\frac{A}{2}e^{st}+\frac{A^*}{2}e^{s^*t}$$  
Let $s=\alpha+j\omega$ and $A=A_m\angle\phi=A_me^{j\phi}$  
  
$$f(t)=\frac{A_me^{j\phi}}{2}e^{(\alpha+j\omega)t}+\frac{A_me^{-j\phi}}{2}e^{(\alpha-j\omega)t}$$  
distribute the exponentials  
  
$$f(t)=\frac{A_m}{2}e^{j\phi}e^{\alpha t}e^{j\omega t}+\frac{A_m}{2}e^{-j\phi}e^{\alpha t}e^{-j\omega t}$$  
pull out the common terms  
  
$$f(t)=A_me^{\alpha t}\left[\frac{e^{j\phi}e^{j\omega t}+e^{-j\phi}e^{-j\omega t}}{2}\right]$$  
combine the exponentials  
  
$$f(t)=A_me^{\alpha t}\left[\frac{e^{j(\omega t+\phi)}+e^{-j(\omega t+\phi)}}{2}\right]$$  
substitute the trigonometric form of cos  
  
$$f(t)=A_me^{\alpha t}cos(\omega t+\phi)$$  
The complex frequency is a real exponential multiplied by a sin/cos wave  
So you can see the complex frequency, $s=\alpha\pm j\omega$,  
describes a sin/cos wave that grows or shrinks exponentially

## 3.1.5 The s-plane

Because the frequency $s$ is now a complex frequency, we can graph it on the complex plane  
  
frequencies that are purely imaginary, represent the sin/cos waves  
while frequencies that are real, represent the growing or shrinking exponentials  
![image7!](./images_s3/image_3_1_7.png)  

If we spin the spin plane so we are looking directly at it, as if it were lying on a table  
then we can view the quadrants  
  
Imaginary frequencies always come as complex conjugate pairs,  
so the top half and the bottom half of the s-plane are always mirror images of each other  
  
* waves with a frequency at the origin are constant DC signals
* waves with a frequency along the positive real axis are growing exponentials
* waves with a frequency along the negative real axis are shrinking exponentials
* waves with a complex conjugate frequencies along the $\pm$ imaginary axis are sin/cos waves
* waves with a complex conjugate frequencies in quadrants I and IV are  growing sin/cos waves
* waves with a complex conjugate frequencies in quadrants II and III are shrinking sin/cos waves
  
The figure below shows the shape of various waves, mapped to the location of their complex frequencies on the s-plane

## 3.1.6 Review of the Laplace Transform

Now you have the whole story, the frequency $s$ is a complex number.  

When we started, we said the Laplace transform converted a time domain signal to a frequency domain signal  
but we implied $s$ was a real number and the Laplace returned a function of $s$, where s was real  
We were holding back the true extent of the frequency variable $s$  
  
Then we applied the Laplace transform to sin/cos waves.  
In this case, we said that the Laplace transform converted the time dependent sin/cos wave to the frequency domain,  
where $s$ was an imaginary number  
and we looked at the frequencies along the $j\omega$ axis.  In fact the Bode plot is always plotted along the $j\omega$ axis  
  
But now we can see that the Laplace transform always converts a time domain signal to the complex frequency domain  
It converts time to the s-plane  
  
Previously, if we made a graph versus frequency,  
we were just showing either the real axis or the imaginary axis  
now we can show the entire complex plane, the s-plane  
  
As an example, the code cell below will plot the transform of the unit step on the s-plane  
  
$$H(s)=\mathcal{L}\left\{u(t)\right\}=\frac{1}{s}$$  
  
Because $s$ is a plane, the magnitude of H(s) must be plotted in 3 dimensions, where H(s) is the z-axis.

## 3.1.7 The Laplace transform creates a function of the complex frequency, s

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

def H_of_s(s):
    H = 1/s
    H_mag = np.absolute(H)
    return H_mag

# Make s-plane
alpha = np.arange(-4.95, 4.95, 0.1)
omega = np.arange(-4.95, 4.95, 0.1)
alpha, omega = np.meshgrid(alpha, omega)
s = alpha + 1j * omega

# Make H(s)
H_mag = H_of_s(s)

# Create plotting objects
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface for magnitude of H(s).
surf = ax.plot_surface(alpha, omega, H_mag, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Label the x and y axes
ax.set(xlabel='real', ylabel='imaginary')

# Customize the z axis.
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
```

You can see, all the energy of the unit step is concentrated near $s=0$  
  
If we take a cross section along the real axis of the s-plane,  
we get the graph we showed you when we introduced the Laplace transform of the unit step

```python
plt.plot(alpha[50],H_mag[50])

plt.xlabel('real axis of s-plane [Np/s]')
plt.ylabel('Magnitude of H(s)')
plt.title('The Laplace transform of the unit step\nalong the real frequency axis')
plt.grid()
```

Recall the Laplace transform of a single real exponential  
The code below shows the plot of the transform of  
  
$$H(s)=\mathcal{L}\left\{e^{-3t}\right\}=\frac{1}{s+3}$$  

```python
def H_of_s(s):
    H = 1/(s+3)
    H_mag = np.absolute(H)
    return H_mag

# Make s-plane
alpha = np.arange(-4.95, 4.95, 0.1)
omega = np.arange(-4.95, 4.95, 0.1)
alpha, omega = np.meshgrid(alpha, omega)
s = alpha + 1j * omega

# Make H(s)
H_mag = H_of_s(s)

# Create plotting objects
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface for magnitude of H(s).
surf = ax.plot_surface(alpha, omega, H_mag, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Label the x and y axes
ax.set(xlabel='real', ylabel='imaginary')

# Customize the z axis.
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
```

Here, the pole moved the frequency $s=-3$ [Np/s], just as expected for a shrinking exponential  
  
for a final example, let's plot the transform of a cos wave  
the cos wave is two exponentials added together  
  
$$H(s)=\mathcal{L}\left\{2cos(3t)\right\}=\mathcal{L}\left\{e^{j3t}+e^{-j3t}\right\}=\frac{1}{s-3j}+\frac{1}{s+3j}$$  

```python
def H_of_s(s):
    H = 1/(s-3j)+1/(s+3j)
    H_mag = np.absolute(H)
    return H_mag

# Make s-plane
alpha = np.arange(-4.95, 4.95, 0.1)
omega = np.arange(-4.95, 4.95, 0.1)
alpha, omega = np.meshgrid(alpha, omega)
s = alpha + 1j * omega

# Make H(s)
H_mag = H_of_s(s)

# Create plotting objects
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface for magnitude of H(s).
surf = ax.plot_surface(alpha, omega, H_mag, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Label the x and y axes
ax.set(xlabel='real', ylabel='imaginary')

# Customize the z axis.
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
```

And now you see the poles at $s=\pm3j$ on the imaginary axis of the complex plane

## 3.1.8 Second order and higher lti systems

We started Fundamentals of Linear Systems by looking at first order systems  
When excited, a first order system responds with a real exponential time dependence  
Passive systems decay exponentially  
The natural frequency of the system sets the decay rate in [Np/s]  

We then looked at first order systems when driven by a sin/cos wave  
We focused on the steady state response, which ignores the natural response of the system,  
and determined the magnitude and phase (Phasor) of the forced response  
However, we did see that a sin/cos wave is really two exponentials added together  
  
First order systems respond with a real exponential at a single real frequency  
Second order systems respond with two exponentials added together  
Because complex numbers come in complex conjugate pairs,  
second order systems can have a complex natural frequency, while first order systems can't  
  
The natural frequencies of a second order system could be two different real frequencies,  
or they could be two repeated real frequencies  
They could also be two complex conjugate frequencies for the natural frequencies of response  
  
For higher order systems, nothing else changes  
A third order system has three natural frequencies  
All three could be real, or one could be real and two could be complex conjugates  
But there are no new types of responses for higher order systems  
As the order increases, lti systems have the same behavior at more and more natural frequencies
