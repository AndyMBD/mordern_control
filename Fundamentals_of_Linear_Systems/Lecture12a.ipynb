## 2.6.1 Review of the exponential wave

The equation for the exponential wave is  
$$Ae^{st}$$  
We have seen that, when excited by a disturbance, a first order linear time invariant system  
will have a response that follows the exponential  
The system will respond at it's natural frequency.  Assume the natural frequency is $s=-a$,  
then then system response, as a function of time, will be  
$$Ae^{-at}$$  
When graphed as a function of time, we get  
![image1!](./images_s2/image_2_6_1.png)  

We can tranform that reponse to the frequency domain using the Laplace transform,  
to get the system response as a function of frequency to be  
$$\frac{A}{s+a}$$  
When the natural response is a negative frequency, the result is a decaying exponential  
Graphing the magnitude (not the sign) of the response, as a function of frequency, gives us  
![image2!](./images_s2/image_2_6_2.png)  

## 2.6.2 Review of the sin/cos wave

The equation for the sin/cos wave is  
$$\frac{A}{2}e^{st}+\frac{A^*}{2}e^{-st}$$  
where the phasor is a complex number with a magnitude and phase angle; $A\;\angle\phi$  
and the frequency, s, is imaginary  
  
The sin/cos wave is two exponentials added together.  
We have not observed a system that has a sin/cos response, because that system would need to be second order  
(A second order system will have a response described by two exponentials added together)  
  
However, we have driven first order systems with a sin/cos wave  
If we choose a specific frequency, like $s=j\omega$,  the equation becomes  
$$\frac{A}{2}e^{j\omega t}+\frac{A^*}{2}e^{-j\omega t}$$  
The graph of the two exponentials with imaginary frequencies added together is  
![image3!](./images_s2/image_2_6_3.png)  

Two exponentials are transformed to the frequency domain in the same way one exponential is transformed  
  
$$\frac{\frac{A}{2}}{s-j\omega}+\frac{\frac{A^*}{2}}{s+j\omega}$$  
When we plot the magnitude (not the phase) in the frequency domain, we will plot it along the imaginary frequency axis  
![image4!](./images_s2/image_2_6_4.png)  
Now you see two poles along the imaginary frequency axis, because of the two exponentials  
One pole has a negative imaginary frequency and the other has a postive imaginary frequency  
Imaginary poles and complex magnitudes are always complex conjugates of each other  
  
So the part of the graph on the negative side of the the imaginary frequencies  
is always a mirror image of the positive side of the imaginary frequencies  
  
It is this imaginary frequency axis, that is the x-axis for the Bode plot  
Because the Bode plot's x-axis is on a log scale, we can't plot the negative imaginary frequencies  
and the result would be a mirror image of the positive imaginary frequencies anyway

## 2.6.3 Periodic waves

A periodic wave is any wave that repeats over time  
Mathematically, a function is periodic in time if $f(t)=f(t+T)$ for all values of t,  
where T is a constant called the period  
[Periodic function](https://en.wikipedia.org/wiki/Periodic_function)  
  
We can apply the test to an exponential wave  
If we assume  
$$f(t)=Ae^{-at}$$  
then  
$$f(t+T)=Ae^{-a(t+T)}$$  
Clearly  
$$Ae^{-at}\neq Ae^{-at}e^{-aT}$$  
if T is a non-zero constant  

We can also apply the test to the sin/cos wave  
Assume  
$$f(t)=\frac{A}{2}e^{j\omega t}+\frac{A^*}{2}e^{-j\omega t}$$  
and  
$$f(t+T)=\frac{A}{2}e^{j\omega(t+T)}+\frac{A^*}{2}e^{-j\omega(t+T)}$$  
If we contrain the period of the wave to the frequency,  
such that $T=\frac{2\pi}{\omega}$, then  

$$f(t+T)=\frac{A}{2}e^{j\omega t}e^{j\omega T}+\frac{A^*}{2}e^{-j\omega t}e^{-j\omega T}$$  
and  
$$f(t+T)=\frac{A}{2}e^{j\omega t}e^{j\omega\frac{2\pi}{\omega}}+\frac{A^*}{2}e^{-j\omega t}e^{-j\omega\frac{2\pi}{\omega}}$$  
which simplfies to  
$$f(t+T)=\frac{A}{2}e^{j\omega t}e^{j2\pi}+\frac{A^*}{2}e^{-j\omega t}e^{-j2\pi}$$  
However, $e^{j2\pi}=e^{-j2\pi}=1$, such that  
$$f(t+T)=f(t)=\frac{A}{2}e^{j\omega t}+\frac{A^*}{2}e^{-j\omega t}$$  
  
so a sin/cos wave is periodic if $T=\frac{2\pi}{\omega}$  
$\omega$ is the frequency in [rad/s], but a cycle requires $2\pi$ radians  
  
A cycle is a repitition of a wave. The number of cycles per second has the units of Hertz, [Hz]  
[Hertz](https://en.wikipedia.org/wiki/Hertz)  
  
If we count the cycles per second, then the frequency is f, in [Hz], rather than $\omega$ in [rad/s]  
The period T, is the amount of time for one cycle, not one radian, so $T=\frac{1}{f}$  
  
There are $2\pi$ radians in a cycle, so $\omega=2\pi f=2\pi\frac{1}{T}$.  This gives us $T=\frac{2\pi}{\omega}$  
A sin/cos wave is periodic

## 2.6.4 An arbitrary periodic wave - the fundamental frequency

It's important to understand, that the waves we are considering now,  
must exist for all time; $-infty<t<+infty$  
The waves in this section are two-sided waves, not one-sided waves as they have been for the Laplace transform  

Any wave that satifies the condition $f(t)=f(t+T)$ for all values of t,  
where T is a constant called the period, is considered a periodic wave  
![image5!](./images_s2/image_2_6_5.png)  
The fundamental frequency is the lowest frequency contained in the wave  
It is found from the period. Once the period is determined,  
the wave will not contain any frequencies lower than the fundamental  
The variable name often used for the fundamental frequncy is $f_o$ in [Hz] or $\omega_o$ in [rad/s]

## 2.6.5 An arbitrary periodic wave - harmonic frequencies

Harmonic frequencies are integer multiples of the fundamental frequency.  
  
Assume a wave has a fundamental frequency of 10 Hz,  
  
the second harmonic is 20 Hz,  
the third harmonic is 30 Hz,  
the fourth harmonic is 40 Hz,  
the fifth harmonic is 50 Hz,  
etc.

## 2.6.6 An example of adding sin waves together

Let's plot a sin wave that has 10 radians in every second  
The sin phasor is an imaginary number, with $-90^\circ$ of phase shift; $A\;\angle-90$  
Let's make the amplitude, $|A|=21$  

```python
import numpy as np
import matplotlib.pyplot as plt

w1 = 10 # [rad/s]
A1 = -21j

t = np.linspace(0, 2, 1000)
y1 = A1*np.exp(1j*w1*t)+np.conjugate(A1)*np.exp(-1j*w1*t)

plt.plot(t, y1.real)
```

Now, let's plot another sin wave with $\omega=30$ [rad/s] and $|A|=7$

```python
w2 = 30 # [rad/s]
A2 = -7j

y2 = A2*np.exp(1j*w2*t)+np.conjugate(A2)*np.exp(-1j*w2*t)

plt.plot(t, y1.real)
plt.plot(t, y2.real)
```

Let's add these two sin waves together

```python
plt.plot(t, (y1+y2).real)
```

You can already see this wave is looking like a square wave!  
  
We are now going to make a plot with our two original sine waves, and add two more  
Here is a summary of the waves we will plot  
  
$\omega=10$ [rad/s] and $A=-21j$  
$\omega=30$ [rad/s] and $A=-7j$  
$\omega=50$ [rad/s] and $A=-4.2j$  
$\omega=70$ [rad/s] and $A=-3j$  

```python
w1 = 10 # [rad/s]
A1 = -21j
w2 = 30 # [rad/s]
A2 = -7j
w3 = 50 # [rad/s]
A3 = -4.2j
w4 = 70 # [rad/s]
A4 = -3j

y1 = A1*np.exp(1j*w1*t)+np.conjugate(A1)*np.exp(-1j*w1*t)
y2 = A2*np.exp(1j*w2*t)+np.conjugate(A2)*np.exp(-1j*w2*t)
y3 = A3*np.exp(1j*w3*t)+np.conjugate(A3)*np.exp(-1j*w3*t)
y4 = A4*np.exp(1j*w4*t)+np.conjugate(A4)*np.exp(-1j*w4*t)

plt.plot(t, y1.real)
plt.plot(t, y2.real)
plt.plot(t, y3.real)
plt.plot(t, y4.real)
```

Finally, let's make a plot with all of these sin waves added together

```python
plt.plot(t, (y1+y2+y3+y4).real)
```

And there we have it!  Four sin waves added together that produce a square wave  
Well, it's almost a square wave.  If we kept adding more and more terms,  
then our square wave would get better and better  
We really need an infinite number of terms to get a perfect square wave  

## 2.6.7 Fourier series

The Fourier series is an infinite series, which is a summation of sin/cos waves  
In the link to the wikipedia page, you should scroll down the animated graphic  
The coefficients for the Fourier series are phasors at different frequencies  
The animated graphic shows the phasors being added together as they spin around the complex plane  
[Fourier series](https://en.wikipedia.org/wiki/Fourier_series)  
  
In equation form, the series is written as  
$$f(t) = c_o + \sum_{n=1}^{\infty} c_ne^{jn\omega_ot} + c_n^*e^{-jn\omega_ot}$$  

where $c_0$, $c_1$, $c_2$, $\cdots$ are phasors, and the coefficients of the sin/cos waves  
The standard divide by 2 part of the phasor, $(\frac{A}{2}\;or\;\frac{A^*}{2})$,  
is incorporated into the Fourier series coefficients  
Therefore Fourier coefficients are really half the magnitude of the sin/cos phasors we defined previously  
  
It turns out, $c_0$, is really the DC bias, or average value of the function  
  
Notice in the equation, that the fundamental frequency $\omega_o$ is multiplied by n,  
thus creating the harmonics of the series  
  
Determining the Fourier series of a function is all about finding the coefficients  

## 2.6.8 You must perform an integral to find the coefficients

The coefficients (including $c_0$) may be found from the following formula  
  
$$c_n = \frac{1}{T}\int_{0}^{T} f(t)e^{-jn\omega_ot} dt$$
  
You integrate the function you want to find the coefficients for, over $\textbf{ONE}$ period.  
You must also multpliy the function by $e^{-jn\omega_ot}$ before you integrate  
  
The code cell below determines the coefficients for our square wave  
using Simpson integration  
[Simpson's rule](https://en.wikipedia.org/wiki/Simpson%27s_rule)

```python
import scipy.signal as sig
import scipy.integrate as integrate

wo = 10 # fundamental frequency [rad/s]
T = 2*np.pi/wo # Period [s]

# create a square wave
t = np.linspace(0, T, 1000)
y = 33*sig.square(10 * t)

# number of terms; add one for c0
num = 7+1
c = [0]*num
for n in range(num):
    f = y * np.exp(-1j * n * wo * t)
    c[n] = (1/T) * integrate.simps(f, t)
    print('c[',n,'] =',c[n])
```

Compare the coefficients with the phasors you chose to create Fourier series of the square wave  
repeated here for convenience  
(There is some rounding error due to numerical integration)  
But notice, the even harmonics are all very close to 0  
and that is why they are not included in the Fourier series of the square wave
  
$\omega=10$ [rad/s] and $A=-21j$  
$\omega=30$ [rad/s] and $A=-7j$  
$\omega=50$ [rad/s] and $A=-4.2j$  
$\omega=70$ [rad/s] and $A=-3j$  

## 2.6.9 Add up the Fourier series just calculated, to verify

```python
# Use a loop to add the sine waves
y = 0
for n in range(num):
    y = y + c[n]*np.exp(1j*n*10*t) + np.conj(c[n])*np.exp(-1j*n*10*t) 
plt.plot(t, y.real)
```

## 2.6.10 Fourier Series, versus Fourier Transform, versus Laplace Transform

The Fourier series represents a conversion of a periodic wave from the time domain to the frequency domain  
The coefficients of the sin/cos waves in the Fourier series, represent the magnitude of impulses in the frequency domian  
The figure below shows the impulses along the frequency axis  
![image6!](./images_s2/image_2_6_6.png)  
The reason the sin waves are transformed to impulses, is because they exist for all time from $-\infty<t<+\infty$  
It's a consequence of the relationship between time and frequency  
Infinite time transforms to a single impulse in frequency, while an impulse in time transforms to all frequencies, $\delta(t)\Rightarrow1$  

The Fourier transform is very similar to the Fourier series, except the waves don't need to be periodic  
and they don't need to exist for all time
If you perform a Fourier transform, rather than calculate the Fourier series, for a square wave,  
how sharp the impulses are, is determined by how much time you sample  
Sample for more time and you will get sharper impulses  
If the transformed wave is not periodic, the resulting frequncies are not pinned to the fundamental and harmonic frequencies  
The figure below shows a typical example of a Fourier transform  
![image7!](./images_s2/image_2_6_7.png)  

We have been quite familiar with the Laplace transform.  We have seen that the Laplace transform  
can convert a time domain signal to the real frequency, $s$, or the imaginary frequency, $s=j\omega$  

We will soon find out that the Laplace transform actually converts a time dependent signal  
to the complex frequency, $s=\alpha+j\omega$, where $s=\alpha$ is the real frequency axis we have been familiar with  
and $s=j\omega$ is the imaginary axis for sin/cos waves we have been studying more recently
